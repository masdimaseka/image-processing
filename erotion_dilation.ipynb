```python
# Path Image
DATA_IMAGE = "/content/drive/MyDrive/dataset-colab/datasetErotionDilation/rm3.jpeg"
```

```python
# import library
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load image
image = cv2.imread(DATA_IMAGE )

#read grey scale image
image_gs=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

#read BW image
thresh,image_bw = cv2.threshold(image_gs,155,255,cv2.THRESH_BINARY)

#defining kernel
kernel = np.ones((5,5),np.uint8)

#getting dialated image
image_dilated=cv2.dilate(image_bw,kernel,anchor=(0, 0),iterations=3)

#getting eroded image
image_eroded=cv2.erode(image_bw,kernel,anchor=(0, 0),iterations=3)

#visualize image
plt.figure(figsize=(15, 5))

plt.subplot(1, 5, 1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 5, 2)
plt.title("Grayscale Image")
plt.imshow(image_gs, cmap='gray')
plt.axis('off')

plt.subplot(1, 5, 3)
plt.title("Binary Image")
plt.imshow(image_bw, cmap='gray')
plt.axis('off')

plt.subplot(1, 5, 4)
plt.title("Dilation Image")
plt.imshow(image_dilated, cmap='gray')
plt.axis('off')

plt.subplot(1, 5, 5)
plt.title("Erotion Image")
plt.imshow(image_eroded, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
