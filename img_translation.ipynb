```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATA_IMAGE = "/content/drive/MyDrive/dataset-colab/translation/ai.jpg"
```

```python
image = cv2.imread(DATA_IMAGE)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
 # Geser ke kanan sejauh 50 piksel dan ke bawah sejauh 100 piksel
tx, ty = 70, 150
```

```python
# Buat matriks translasi
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])

# Terapkan translasi
translated_image = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0]))

# Convert hasil translasi ke RGB
translated_image_rgb = cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB)

# Visualisasi hasil
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image_rgb)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Translated Image")
plt.imshow(translated_image_rgb)
plt.axis("off")
plt.show()
```

```python
output_path = "translated_image.jpg"
cv2.imwrite(output_path, translated_image)
print(f"Hasil translasi disimpan di: {output_path}")
```
