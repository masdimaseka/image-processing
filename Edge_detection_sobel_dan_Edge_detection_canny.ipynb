```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATA_IMAGE = "/content/drive/MyDrive/dataset-colab/edge/globe.jpeg"
img = cv2.imread(DATA_IMAGE, cv2.IMREAD_GRAYSCALE)
```

```python
# Edge Detection dengan Sobel
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)  # Derivatif X
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)  # Derivatif Y
sobel_combined = cv2.magnitude(sobelx, sobely)     # Kombinasi magnitude

# Konversi hasil Sobel ke tipe uint8
sobel_combined = cv2.convertScaleAbs(sobel_combined)

# Edge Detection dengan Canny
canny = cv2.Canny(img, 150, 250)  # Threshold

# Tampilkan hasil
plt.figure(figsize=(10, 6))
plt.subplot(1, 3, 1), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.axis('off')

plt.subplot(1, 3, 2), plt.imshow(sobel_combined, cmap='gray'), plt.title('Sobel Edge')
plt.axis('off')

plt.subplot(1, 3, 3), plt.imshow(canny, cmap='gray'), plt.title('Canny Edge')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
output_path = "edge_sobel.jpg"
cv2.imwrite(output_path, sobel_combined)
print(f"Hasil translasi disimpan di: {output_path}")

output_path = "edge_canny.jpg"
cv2.imwrite(output_path, canny)
print(f"Hasil translasi disimpan di: {output_path}")
```
