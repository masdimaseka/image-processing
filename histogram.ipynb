```python
# Menyimpan gambar pada sebuah variable
images = ["/content/drive/MyDrive/dataset-colab/SampleGambarHisto/gambar-1.jpg",
              "/content/drive/MyDrive/dataset-colab/SampleGambarHisto/gambar-2.jpeg",
              "/content/drive/MyDrive/dataset-colab/SampleGambarHisto/gambar-3.jpeg" ]
```

```python
# Membuat function untuk mengetahui histrogram grayscale
def getHistogramGrayscale(image_path, index):

  # Membaca gambar dan merubah gambar menjadi grayscale
  image = cv2.imread(image_path)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  # Menghitung histogram
  hist = cv2.calcHist([image], [0], None, [256], [0, 256])

  plt.figure(figsize=(10, 5))

  # Menampilkan Gambar
  plt.subplot(1, 2, 1)
  plt.imshow(image, cmap="gray")
  plt.title(f"Gambar {index + 1}")
  plt.axis("off")

  # Menampilkan hasil histogram
  plt.subplot(1, 2, 2)
  plt.plot(hist, color="gray")
  plt.xlim([0, 256])
  plt.title('Histogram Grayscale')

  plt.show()
```

```python
# Membuat function untuk mengetahui histrogram RGB
def getHistogramRGB(image_path, index):

  # Membaca gambar dan merubah gambar menjadi RGB
  image = cv2.imread(image_path)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

  # Memisahkan gambar menjadi 3 channel terpisah: Red, Green, dan Blue.
  r, g, b = cv2.split(image)

  # Menghitung histogram Red, Gren, dan Blue
  hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])
  hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
  hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])

  plt.figure(figsize=(20, 5))

  # Menampilkan Gambar
  plt.subplot(1, 4, 1)
  plt.imshow(image)
  plt.title(f"Gambar {index + 1}")
  plt.axis("off")

  # Menampilkan hasil histogram Red
  plt.subplot(1, 4, 2)
  plt.plot(hist_r, color="red")
  plt.xlim([0, 256])
  plt.title('Histogram Red')

  # Menampilkan hasil histogram Green
  plt.subplot(1, 4, 3)
  plt.plot(hist_g, color="green")
  plt.xlim([0, 256])
  plt.title('Histogram Green')

  # Menampilkan hasil histogram Blue
  plt.subplot(1, 4, 4)
  plt.plot(hist_b, color="blue")
  plt.xlim([0, 256])
  plt.title('Histogram Blue')

  plt.show()
```

```python
# Menjalan function getHistogramGrayscale dan getHistogramRGB

print("Histogram pada Gambar Grayscale")
for i in range(len(images)):
    getHistogramGrayscale(images[i], i)

print("Histogram pada Gambar RGB")
for i in range(len(images)):
    getHistogramRGB(images[i], i)
```
